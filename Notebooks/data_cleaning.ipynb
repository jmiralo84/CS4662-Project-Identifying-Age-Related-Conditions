# All Imports & Loading Data

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

train = pd.read_csv(r"../Data/train.csv")

print(train.columns.tolist())
print(train.shape)
display(train.head())
print(train.info())
````

````python
test = pd.read_csv(r"../Data/test.csv")

print(test.columns.tolist())
print(test.shape)
display(test.head())
print(test.info())
````

# Cleaning Data

### Column Check
Checking whether the train and test dataset have the same feature columns, also compare the column difference between them.

````python
train_cols = set(train.columns)
test_cols = set(test.columns)

train_only = train_cols - test_cols
test_only = test_cols - train_cols

id_col = "Id"
label_col = "Class"

print("Columns only in train:", train_only if train_only else None)
print("Columns only in test:", test_only if test_only else None)
print("Id in train:", id_col in train.columns, "| Id in test:", id_col in test.columns)
print("Label (Class) in train:", label_col in train.columns)
````

### Optional: Load greeks.csv

The competition also provides an additional file called `greeks.csv`.

Supplemental metadata, only available for the training set:
- ```Alpha``` Identifies the type of age-related condition, if present.
    - ```A``` No age-related condition. Corresponds to class 0.
    - ```B, D, G``` The three age-related conditions. Correspond to class 1.
- ```Beta, Gamma, Delta``` Three experimental characteristics.
- ```Epsilon``` The date the data for this subject was collected. Note that all of the data in the test set was collected after the training set was collected.

(This step is optional and safe to skip if not needed.)

````python
greeks = pd.read_csv(r"../Data/greeks.csv")
print(greeks.columns.tolist())
print(greeks.shape)
display(greeks.head())
````

````python
print("train before merge: ", train.shape)
train = train.merge(greeks, on="Id", how="left")
print("train after merge: ", train.shape)
````

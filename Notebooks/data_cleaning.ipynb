# All Imports & Loading Data

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import seaborn as sns

train = pd.read_csv(r"../Data/train.csv")

print(train.columns.tolist())
print(train.shape)
display(train.head())
print(train.info())
````

````python
test = pd.read_csv(r"../Data/test.csv")

print(test.columns.tolist())
print(test.shape)
display(test.head())
print(test.info())
````

# Cleaning Data

### Column Check
Checking whether the train and test dataset have the same feature columns, also compare the column difference between them.

````python
train_cols = set(train.columns)
test_cols = set(test.columns)

train_only = train_cols - test_cols
test_only = test_cols - train_cols

id_col = "Id"
label_col = "Class"

print("Columns only in train:", train_only if train_only else None)
print("Columns only in test:", test_only if test_only else None)
print("Id in train:", id_col in train.columns, "| Id in test:", id_col in test.columns)
print("Label (Class) in train:", label_col in train.columns)
````

### Optional: Load greeks.csv

The competition also provides an additional file called `greeks.csv`.

Supplemental metadata, only available for the training set:
- ```Alpha``` Identifies the type of age-related condition, if present.
    - ```A``` No age-related condition. Corresponds to class 0.
    - ```B, D, G``` The three age-related conditions. Correspond to class 1.
- ```Beta, Gamma, Delta``` Three experimental characteristics.
- ```Epsilon``` The date the data for this subject was collected. Note that all of the data in the test set was collected after the training set was collected.

(This step is optional and safe to skip if not needed.)

````python
greeks = pd.read_csv(r"../Data/greeks.csv")
print(greeks.columns.tolist())
print(greeks.shape)
display(greeks.head())
````

````python
print("train before merge: ", train.shape)
train = train.merge(greeks, on="Id", how="left")
print("train after merge: ", train.shape)
````

### Check for Duplicates and Missing Values

1. Removes exact duplicate rows (if any exist)  
2. Computes the percentage of missing values per column  
3. Displays the top columns with missing data  

Since the dataset is small, we only remove exact duplicates and do not drop rows due to missing values.

````python
# remove exact duplicates
before = train.shape[0]
train = train.drop_duplicates()
after = train.shape[0]
print("Dropped duplicates:", before - after)

# missingness summary
missing_perc = train.isna().mean().sort_values(ascending=False)
missing_perc = missing_perc[missing_perc > 0]

print("Columns with missing values:", len(missing_perc))
display((missing_perc * 100).round(4).head(20).to_frame("missing %"))
````

### Fill Missing Values and Align Train/Test Columns

Missing values are handled using statistics computed from the training set:

- Numeric columns → filled with median  
- Categorical columns → filled with mode  

After filling, we align train and test to ensure they have identical feature columns before modeling.

````python
# create a list of all columns except the label
feature_cols = [col for col in train.columns if col != label_col]

# loop through each feature column and fill missing values
for col in feature_cols:
    
    # exclude label_col if not done so
    if col == label_col:
        continue

    # if the column is numeric, fill missing values with the median
    if train[col].dtype != "object":
        median = train[col].median()
        train[col] = train[col].fillna(median)

        # use the same median from train to fill missing values in test
        if col in test.columns:
            test[col] = test[col].fillna(median)

    # if the column is categorical/object, fill missing values with the mode
    else:
        mode = train[col].mode()
        mode = mode[0] if len(mode) > 0 else None
        train[col] = train[col].fillna(mode)

        # use the same mode from train to fill missing values in test
        if col in test.columns:
            test[col] = test[col].fillna(mode)

# remove the label temporarily so train can be aligned with test
train_no_label = train.drop(columns=[label_col])

# align train and test to make sure they have the same feature columns
# join="left" keeps the train feature structure
# fill_value=0 fills any missing columns created during alignment
train_no_label, test = train_no_label.align(test, join="left", axis=1, fill_value=0)

# add the label column back to train
train = pd.concat([train_no_label, train[[label_col]]], axis=1)

print("Aligned shapes -> train:", train.shape, "test:", test.shape)
````

````python
# check missing values after cleaning
train_missing_after = train.isna().sum().sum()
test_missing_after = test.isna().sum().sum()

print("Total missing values in cleaned train:", train_missing_after)
print("Total missing values in cleaned test:", test_missing_after)
````

### Save Cleaned Train and Test Files

After preprocessing, we save cleaned versions of both datasets.

- train.csv -> clean_train.csv
- test.csv -> clean_test.csv

````python
out_dir = Path("../Data")
out_dir.mkdir(parents=True, exist_ok=True)

train.to_csv(out_dir / "clean_train.csv", index=False)
test.to_csv(out_dir / "clean_test.csv", index=False)

print("Saved:")
print(out_dir / "clean_train.csv")
print(out_dir / "clean_test.csv")
````

````python
# loading cleaned data
clean_train = pd.read_csv(r"../Data/clean_train.csv")
clean_test = pd.read_csv(r"../Data/clean_test.csv")

print("loading cleaned data...")

print(clean_train.shape)
print(clean_test.shape)

display(clean_train.head())
display(clean_test.head())
````

### Plots

- How many samples belong to each class.
- Histograms display unusual feature distributions
- Boxplots for detecting extreme values and spread of the data
- Correlation heatmap to show correlations among numeric features

````python
plt.figure(figsize=(6, 4))
clean_train[label_col].value_counts().sort_index().plot(kind="bar")
plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
````

````python
# Plot histograms

# select numeric columns
numeric_cols = clean_train.select_dtypes(include=[np.number]).columns.tolist()

# remove Id and label from histogram plotting
numeric_cols = [col for col in numeric_cols if col not in [id_col, label_col]]

# limit to show 10 cols for readability
show_cols = numeric_cols[:10]

print("Plotting histograms for:", show_cols)

for col in show_cols: 
    plt.figure(figsize=(6, 4))
    clean_train[col].hist(bins=30)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
````

````python
# boxplots for a subset of numeric columns

print("Plotting boxplots for:", show_cols)

for col in show_cols:
    plt.figure(figsize=(6, 4))
    plt.boxplot(clean_train[col].dropna())
    plt.title(f"Boxplot of {col}")
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
````

````python
# correlation heatmap

corr_matrix = clean_train[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=False)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
````
